```python
path="F:/IT Field/PHONE_PAY/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path)
#Agg_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

clm={'State':[], 'Year':[],'Quater':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in Agg_state_list:
    p_i=path+i+"/"            #F:/IT Field/PHONE_PAY/data/aggregated/transaction/country/india/state/andaman
    Agg_yr=os.listdir(p_i)

    for j in Agg_yr:
        p_j=p_i+j+"/"         #F:/IT Field/PHONE_PAY/data/aggregated/transaction/country/india/state/andaman/2018
        Agg_yr_list=os.listdir(p_j)

        for k in Agg_yr_list:
            p_k=p_j+k         #F:/IT Field/PHONE_PAY/data/aggregated/transaction/country/india/state/andaman/2018/1
            Data=open(p_k,'r')  #To open the json file 
            D=json.load(Data)

            for z in D['data']['transactionData']:
                Name=z['name']
                count=z['paymentInstruments'][0]['count']
                amount=z['paymentInstruments'][0]['amount']
                clm['Transaction_type'].append(Name)
                clm['Transaction_count'].append(count)
                clm['Transaction_amount'].append(amount)
                clm['State'].append(i)
                clm['Year'].append(j)
                clm['Quater'].append(int(k.strip('.json')))

#Succesfully created a dataframe
Agg_Trans=pd.DataFrame(clm)
```

```python
path1="F:/IT Field/PHONE_PAY/data/aggregated/user/country/india/state/"
user_state_list=os.listdir(path1)

clm1={'State':[], 'Year':[],'Quater':[],'User_brand':[], 'User_count':[], 'User_percentage':[]}

for i in user_state_list:
    p_i=path1+i+"/"
    user_yr=os.listdir(p_i)

    for j in user_yr:
        p_j=p_i+j+"/"
        user_yr_list=os.listdir(p_j)
        
        for k in user_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            E=json.load(Data)

            try:
                for z in E['data']['usersByDevice']:
                    Brand=z['brand']
                    count=z['count']
                    amount=z['percentage']
                    clm1['User_brand'].append(Brand)
                    clm1['User_count'].append(count)
                    clm1['User_percentage'].append(amount)
                    clm1['State'].append(i)
                    clm1['Year'].append(j)
                    clm1['Quater'].append(int(k.strip('.json')))
            except:
                pass

#Succesfully created a dataframe

Agg_user=pd.DataFrame(clm1)

Agg_user['State']=Agg_user['State'].str.title()
Agg_user['State']=Agg_user['State'].replace('andaman-&-nicobar-islands','Andaman & Nicobar')
Agg_user['State']=Agg_user['State'].str.replace('-',' ')
Agg_user['State']=Agg_user['State'].replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
```

```python
path3="F:/IT Field/PHONE_PAY/data/map/user/hover/country/india/state/"
map_user_state_list=os.listdir(path3)

clm3={'State':[], 'Year':[],'Quater':[],'Districts':[],'Registered_Users':[], 'App_Opens':[]}

for i in map_user_state_list:
    p_i=path3+i+"/"
    map_user_yr=os.listdir(p_i)

    for j in map_user_yr:
        p_j=p_i+j+"/"
        map_user_yr_list=os.listdir(p_j)
        
        for k in map_user_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            F=json.load(Data)

            for z in F['data']['hoverData'].items():
                District=z[0]
                RegisteredUsers=z[1]['registeredUsers']
                AppOpens=z[1]['appOpens']
                clm3['Districts'].append(District)
                clm3['Registered_Users'].append(RegisteredUsers)
                clm3['App_Opens'].append(AppOpens)
                clm3['State'].append(i)
                clm3['Year'].append(j)
                clm3['Quater'].append(int(k.strip('.json')))

Map_User=pd.DataFrame(clm3)

Map_User['State']=Map_User['State'].str.title()
Map_User['State']=Map_User['State'].replace('andaman-&-nicobar-islands','Andaman & Nicobar')
Map_User['State']=Map_User['State'].str.replace('-',' ')
Map_User['State']=Map_User['State'].replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
```

```python
path4="F:/IT Field/PHONE_PAY/data/top/transaction/country/india/state/"
top_trans_state_list=os.listdir(path4)

clm4={'State':[], 'Year':[],'Quater':[],'Pincode':[],'Transaction_count':[], 'Transaction_amount':[]}

for i in top_trans_state_list:
    p_i=path4+i+"/"
    top_trans_yr=os.listdir(p_i)

    for j in top_trans_yr:
        p_j=p_i+j+"/"
        top_trans_yr_list=os.listdir(p_j)
        
        for k in top_trans_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            G=json.load(Data)

            for z in G['data']['pincodes']:
                    pincode=z['entityName']
                    count=z['metric']['count']
                    amount=z['metric']['amount']
                    clm4['Pincode'].append(pincode)
                    clm4['Transaction_count'].append(count)
                    clm4['Transaction_amount'].append(amount)
                    clm4['State'].append(i)
                    clm4['Year'].append(j)
                    clm4['Quater'].append(int(k.strip('.json')))

Top_Trans=pd.DataFrame(clm4)

Top_Trans['State']=Top_Trans['State'].str.title()
Top_Trans['State']=Top_Trans['State'].replace('andaman-&-nicobar-islands','Andaman & Nicobar')
Top_Trans['State']=Top_Trans['State'].str.replace('-',' ')
Top_Trans['State']=Top_Trans['State'].replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
```
